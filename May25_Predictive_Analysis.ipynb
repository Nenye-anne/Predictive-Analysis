### Correlation plot (Scatter plot and heatmap)

#### Scatter Plot
Relationship between two different variables (columns). The relationship can be either:
* Positive (upward trend)
* Negative(downward trend)
* No relationship (no trend)

```python
sns.scatterplot(data=iris_df, x='sepal length (cm)', y='petal length (cm)')
plt.title("Relationshp Between Sepal Length and Petal Lenght")
plt.xlabel("Sepal Lenght")
plt.ylabel('Petal Length')
plt.show()
```

There seems to be an upward trend trend for higher values of sepal length and petal lenght.

Lets view the relationship among the features with a single plot. This makes sense because we want to ensure there is no multicolinearity among the feature variables to avoid error when machine learning algorithm is learning from the dataset:

```python
sns.pairplot(iris_df[iris_feature_names])
plt.show()
```

```python
iris_df.columns
```

```python
sns.scatterplot(data=iris_df, x='sepal length (cm)', y='petal length (cm)', hue='Species')
plt.title("Relationshp Between Sepal Length and Petal Lenght")
plt.xlabel("Sepal Lenght")
plt.ylabel('Petal Length')
plt.show()
```

#### Heatmap

This is the numerical representation of relationship unlike scatter plot which only shows patterns.

Correlation values is between -1 and 1
* Positive correlation/relationship: values close 1
* Negative correlation/relationship: values closer to -1
* No correlation/relationship: values around 0.

Using pandas `.corr()` method:

```python
#  we first need calculate the correlation value
correlation_values = iris_df[iris_feature_names].corr()
correlation_values
```

```python
# to visualize the correlation values
sns.heatmap(correlation_values, annot=True, cmap="Reds")
plt.show()
```

### Seperating the Features from Target

```python
iris_df
```

```python
iris_df.columns
```

```python
# features is X variable
feature_data = iris_df[['sepal length (cm)',
                        'sepal width (cm)',
                        'petal length (cm)',
                        'petal width (cm)']]


feature_data.head()
```

```python
# target: called y variable
target = iris_df.target

target.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
predictor = LogisticRegression()
```

```python
# algorithm will learn pattern fromm features
predictor.fit(feature_data, target)
```

```python
predicted_target = predictor.predict(feature_data)
```

```python
result_df = pd.DataFrame({'Actual Target': target,
                          "Predicted Target": predicted_target})
```

```python
result_df
```

# Evaluating the Model

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
result = confusion_matrix(target, predicted_target)
result
```

```python
iris_target_names
```

```python

confusion_matrix_df = pd.DataFrame(data=result, columns=iris_target_names)
confusion_matrix_df.index = iris_target_names
confusion_matrix_df
```

```python
print(classification_report(target, predicted_target))
```
